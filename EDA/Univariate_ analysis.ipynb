```python
for col in titanic.columns:
    # Check if the column is of numeric type (integer or float)
    if titanic[col].dtype in [int, float]:
        # Create a box plot for the current column
        sns.boxplot(x=titanic[col], data=titanic)
        # Add title for better understanding
        plt.title(f'Box Plot for {col}')
        
        # Show the plot
        plt.show()
```
