### SIMPLE PROGRAM TO PREDICT USING LINEAR REGRESSION

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Sample data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([1,4,9,16,25])
```

```python
# Create and train the model
model = LinearRegression()
model.fit(X, y)
```

```python
# Coefficients
print("Intercept:", model.intercept_)
print("Coefficient:", model.coef_)

```

```python
# Make predictions
X_new = np.array([[7]])
predictions = model.predict(X_new)
print(predictions)
```

```python
# Generate line of best fit
X_fit = np.linspace(1, 7, 50).reshape(-1, 1)
y_fit = model.predict(X_fit)
```

```python
# Plotting the data points and the best fit line
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X_fit, y_fit, color='red', label='Best Fit Line')
plt.scatter(X_new, predictions, color='green', label='Prediction for X=7')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Best Fit Line using Linear Regression')
plt.legend()
plt.show()
```
